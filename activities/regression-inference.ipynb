 6. Below we are going to estimate common values for the slope coefficient through resampling/bootstrapping. Inside the `resample_baby` function below, fill in the appropriate model for the linear model line of code (make sure this matches from your equation above). Is the slope estimate from running the resample code below once the same as the estimate above? Why is the same or different?

```r
resample_baby <- function(...) {
  baby_resample <- baby %>%
    sample_n(nrow(baby), replace = TRUE)

  baby_resample %>%
    lm(!! ~ @@, data = .) %>%
    coef(.) %>%
    .[2] %>%
    data.frame()
}

resample_baby()
```

 7. Now that there is a function that does steps 1 - 3, these processes can now be repeated many times. In place of the "%%" below, add the number of times to resample and get a slope estimate. How do changes in the number of times the data are resampled impact the results?

```r
baby_coef <- map(1:%%, resample_baby) %>%
  bind_rows()
names(baby_coef) <- 'slope'
```

 8. Below is code to visualize a density curve from the results above, summarize the distribution of slopes in a few sentences.

```r
gf_density(~ slope, data = baby_coef)
```
