 # True Experiments - AB Testing

 This section of notes is going to highlight a true experiment to compare and contrast with the natural experiments shown in the first week of the course. The type of true experiment we are going to talk about today are popular in business, particularly within the technology sector and are referred to as AB tests.

 In an AB test, only one element is altered, while everything else remains the same. For example, if you think about a webpage, perhaps the font could be changed or the location of the button is moved from the bottom right to the bottom left. Everything else on the web page would stay the same. Then individuals who go to the web page are randomly assigned to get the old page versus the new page with the single alteration.

 ## Real-World AB Test Example

 In this example, we will explore some real-world data from a video game called Cookie Cats. This game is one where you match 3 or more similar shape/colors of cookies that are fed to cats or dog bosses to progress through the levels. It is similar to the bejeweled type of games that have become popular, you can watch a [trailer for the game](https://youtu.be/GaP5f0jVTWE). The game is a free-to-play style game where is it free to download, but you can only progress so far while playing for free. There are certain gates that stop your progress that you can either pay to remove the gate or you have to wait so long to continue playing. The experiment in this situation for which we have some data is looking at changing when players encounter this first gate, either after 30 or 40 levels of the game.

 ## Cookie Cats Data

 Let's now import the data and look at the data.

```r
.libPaths("../RPackages")

library(tidyverse)
library(ggformula)
library(mosaic)

theme_set(theme_bw(base_size = 16))

cookie_cats <- read_csv("https://raw.githubusercontent.com/lebebr01/psqf_4143/main/data/cookie_cats.csv")

head(cookie_cats)
```

 As you can see, there are a handful of attributes in the data. The first one is an ID for the player. The second one is the version of the game that the player received, namely, whether the gate was after 30 or 40 levels. The following three attributes are ones that could be considered outcomes. One is the total number of game rounds that a player has played and the other two are whether the player returns after 1 day or 7 days.

 I'm going to focus on the retention attributes for now, but we could also explore the number of total games that a player plays. The retention could be more useful for the developer as they likely would get advertisement revenue from players playing over a longer period of time.

```r
gf_bar(~ version, fill =  ~ retention_1, data = cookie_cats, position = 'fill')
```

```r
gf_bar(~ version, fill =  ~ retention_7, data = cookie_cats, position = 'fill')
```

 These two figures show bar graphs for each version of the game on the x-axis. The two colors on the x-axis represent if a player came back to the game a day later or 7 days later respectively. The FALSE categories (red) are players that did not return and TRUE categories (blue) are players that did return. There are very subtle differences when looking at the two columns suggesting no big differences, however in the 7 day retention, the gate at level 30 appears to be a bit higher and so placing the gate earlier may improve player retention. Retaining a couple additional percentages of players could have a large impact on the developers revenue for games with many players.
