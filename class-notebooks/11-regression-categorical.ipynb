   ##Description of the Data
 These data are from a bike share program in Minneapolis Minnesota called Nice Ride. These programs have become popular in larger metros and allow individuals to rent a bike for a short period of time at specified stations. Users return the bikes to the same station or another station in a different part of the city and are charged per ride based on how long the bike was rented for. The pricing structure for this program can be found here: https://www.niceridemn.com/pricing. The gist of the pricing structure is that users can pay for each individual ride where every 30 minutes of bike renting time costs $2 or you can buy a day pass for unlimited 30 minute rides for 24 hours. There is also an option for locals to buy an annual membership which comes with unlimited 60 minute rides for an entire year. Also, if you are interested, a real time map of the bike stations and bikes that are available can be seen here: https://member.niceridemn.com/map/.

 The data was downloaded directly from the Nice Ride website and contain individual trip data for the month of June 2019. Nice Ride releases trips that are longer than 60 seconds and some additional processing was done to remove rides that were longer than 2 hours (120 minutes).  The data was downloaded directly from the Nice Ride website and contain individual trip data for the month of June 2019. Nice Ride releases trips that are longer than 60 seconds and some additional processing was done to remove rides that were longer than 2 hours (120 minutes). The data contain the following variables, note the variable name is in bold with a short description of what the variable represents:
 + **tripduration**: length of bike rental in seconds
 + **start_station_id**: ID of the station where the bike was rented
 + **start_station_name**: Name of the station where the bike was rented
 + **end_station_id**: ID of the station where the bike was returned to
 + **end_station_name**: Name of the station where the bike was returned to
 + **bikeid**: The unique ID of the specific bike rented
 + **usertype**: The type of rental user, "Customer" is a per bike rental, "Subscriber" is an annual Nice Ride subscriber
 + **birth_year**: The birth year of the individual renting the bike
 + **gender**: The gender of the individual renting the bike, 0 = unknown; 1 = male; 2 = female
 + **bike_type**: The type of bike rented, "Classic" is the bike from a station; "Dockless" is a bike that floats around and does not need to start or return from a station.
 + **tripduration_minutes**: The length the bike was rented in minutes, i.e. the varible tripduration / 60.
 + **more_than_30minutes**: Whether the rental was 30 minutes or longer.
 + **more_than_60minutes**: Whether the rental was 60 minutes or longer.
