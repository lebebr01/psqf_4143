More detail on the specific variables in the data are shown below. For data that represent categories, the categories are listed for reference. Please do not hesitate if you have questions regarding the data.

 - **age**: age of individual, rounded to whole years
 - **workclass**: The class of job that the individual has; *Private* = works for a private companty, *government* = local, state, or federal government employee, *self-employ* = self-employed.

 - **sex**: Sex; Female, Male
 - **hours_per_week**: Number of hours worked per week on average.
 - **native_country**: Country employed in, this is only United-States

 ## Question 7
 Complete the following code to split the data into the training and test data. Fill in the amount of data to be included in the training data in place of the "^^" in the code below.

```python
set.seed(2019)
income_split <- initial_split(income, prop = ^^)
income_train <- training(income_split)
income_test <- testing(income_split)
```

 ## Question 8
 Complete the following code to fit a classification model that predicts whether an individual makes more than \\$50000. Fill in the data attribute of interest in place of "^^" in the code below. Ensure that the attributes placed inside "%%" are separated by the `+` sign.

```python
class_tree <- rpart(^^ ~ age + workclass + education + marital_status + occupation + race + sex + hours_per_week, 
   method = 'class', data = income_train)

rpart.plot(class_tree, roundint = FALSE, type = 3, branch = .3)
rpart.rules(class_tree)
```

 ## Question 9
 Complete the evaluate the classification accuracy of the model on the test data. Fill in the data attribute of interest in place of "^^" in the code below and add the additional attributes in place of "%%" below that predict the outcome. Ensure that the attributes placed inside "%%" are separated by the `+` sign.

```python
income_predict_test <- income_test %>%
  mutate(tree_predict = predict(class_tree, newdata = income_test, type = 'class'))
income_predict_test %>%
  mutate(same_class = ifelse(greater_than_50k == tree_predict, 1, 0)) %>%
  df_stats(~ same_class, mean, sum, length)
```

 ## Question 10
 Complete the following code to fit another classification model that predicts whether an individual makes more than \\$50000. Fill in the data attribute of interest in place of "^^" in the code below. Ensure that the attributes placed inside "%%" are separated by the `+` sign.
