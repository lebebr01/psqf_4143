 ## Income data
 Income data from the United States will be used for this take home exam. The data contains individual information about the income, type of job, age, race, sex, and other characteristics from 1994 and published in 1996. This data is a subsample of a larger pool of data obtained here: [UC Irvine Data Repo](https://archive.ics.uci.edu/ml/datasets/Adult). More detail on the specific variables in the data are shown below. For data that represent categories, the categories are listed for reference. Please do not hesitate if you have questions regarding the data.

 - **age**: age of individual, rounded to whole years
 - **workclass**: The class of job that the individual has; *Private* = works for a private company, *government* = local, state, or federal government employee, *self-employ* = self-employed.
 - **education**: Highest education completed; *less-HS* = did not complete high school, *HS-grad* = received high school diploma, *Some-college* = completed some college, *Assoc* = completed an associated degree, *Masters* = completed a masters degree, *Doctorate* = completed a PhD degree, *Prof-school* = completed a professional school.
 - **marital_status**: Marital status; Divorced; Married; Never-married; Separated; Widowed
 - **occupation**: Occupation of employment; Adm-clerical, Craft-repair, Exec-managerial, Farming-fishing, Handlers-cleaners, Machine-op-inspct, Other-service, Prof-specialty, Protective-serv, Sales, Tech-support, Transport-moving
 - **race**: Race; Black, White
 - **sex**: Sex; Female, Male
 - **hours_per_week**: Number of hours worked per week on average.
 - **native_country**: Country employed in, this is only United-States
 - **greater_than_50k**: Dichotomous income variable; 50000 dollars or more vs less than or equal to 50000 dollars.
 - **greater_than_50k_num**: Dichotomous income variable where 1 = 50000 dollars or more, 0 = less than or equal to 50000 dollars

 ## Question 8
 Complete the following code to fit a classification model that predicts whether an individual makes more than 50000 dollars.

```r
class_tree <- rpart(greater_than_50k ~ age + workclass + education + marital_status + occupation + race + sex + hours_per_week, 
   method = 'class', data = income_train)

rpart.plot(class_tree, roundint = FALSE, type = 3, branch = .3)
rpart.rules(class_tree)
```
