 ## Life Expectancy data
 These data come from the Global Health Observatory which is part of the World Health Organization. The data were obtained from [kaggle](https://www.kaggle.com/kumarajarshi/life-expectancy-who).

 - **Country**: Country in world
 - **Year**: Year of data recorded; ranges from 2000 to 2015
 - **Status**: Country status; Developed or Developing
 - **Life_expectancy**: Life Expectancy in age
 - **adult_mortality**: Adult mortality rates of both sexes (probability of dying between 15 and 60 year per 1000 population)
 - **infant_deaths**: Number of infant deaths per 1000
 - **alcohol**: Alcohol consumption rates per capita (in liters of pure alcohol)
 - **percentage_expend**: Percentage of gross domestic product per capita expenditure on health
 - **hepatitis_b_immunization**: Rate of hepatitis B immunization for 1 year olds
 - **measles_rate**: Number of reported measles cases per 1000
 - **bmi**: Average body mass index of country
 - **under_five_deaths**: Number of under-five year old deaths per 1000
 - **polio_immunization**: Rate of polio immunization for 1 year olds
 - **total_expend**: Percentage of government expenditure on health
 - **diphtheria_immunization**: Diphtheria (DTP3) immunization for 1 year olds
 - **HIV_AIDS_deaths**: Deaths per 1000 live births of HIV/AIDS for 0 to 4 year olds
 - **gdp**: Gross Domestic Product per capita (in USD)
 - **population**: Population of the country
 - **thinness_1019**: Prevalence of thinness ages 10 to 19, reported as a percentage
 - **thinness_59**: Prevalence of thinness ages 5 to 9, reported as a percentage
 - **income_composition**: Human Development Index in terms of income composition of resources (index ranging from 0 to 1)
 - **schooling**: Number of years of schooling (years)

 ## Question 8
 Complete the code below to fit a linear model that predicts a country's life expectancy using the development status of the country. Fill in the outcome attribute in place of "^^" in the code below and add the development status attribute in place of "@@" **1 pt**
 *Note: This code may take up to a minute to run.*

```python
resample_status <- function(...) {
  le_resample <- life_expectancy %>%
    sample_n(nrow(life_expectancy), replace = TRUE)

  le_resample %>%
    lm(^^ ~ @@, data = .) %>%
    coef(.) %>%
    .[2] %>%
    data.frame()
}
status_coef <- map(1:10000, resample_status) %>%
  bind_rows()
names(status_coef) <- 'slope'

gf_density(~ slope, data = status_coef)
```

 ## Question 10
 Suppose instead of fitting the linear model above, the median differences between the developing status of a country was calculated.
 *Note: This code could take up to 60 seconds to complete*

```python
resample_le_median <- function(...) {
  le_resample <- life_expectancy %>%
    sample_n(nrow(life_expectancy), replace = TRUE)

  med_est <- le_resample %>%
    df_stats(Life_expec ~ Status, median) %>%
    spread(Status, median_Life_expec)
  
  med_est %>% 
    mutate(developing_developed = Developing - Developed)
}

le_median <- map(1:10000, resample_le_median) %>%
  bind_rows()

gf_density(~ developing_developed, data = le_median)
```
